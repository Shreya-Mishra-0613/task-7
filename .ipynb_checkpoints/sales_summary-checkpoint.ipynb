```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Connecting with database
connect = sqlite3.connect("sales_data.db")
```

```python
# SQL query
query = """
SELECT 
    product, 
    SUM(quantity) AS total_quantity, 
    SUM(quantity * price) AS revenue
FROM sales
GROUP BY product;
"""
```

```python
# Save query results in a Pandas DataFrame
df = pd.read_sql_query(query, connect)
```

```python
#Previewing results
df
```

```python
df.plot(kind='bar', x='product', y='revenue', legend=False, color="purple")
plt.title("Revenue by Product")
plt.xlabel("Product")
plt.ylabel("Revenue")
plt.tight_layout()
plt.savefig("sales_chart.png")
plt.show()
```

#### Insights
- Laptops have generated the maximum revenue
- Mouse, Webcam and Keyboard generated the least revenues

```python
df.plot(kind='bar', x='product', y='total_quantity', legend=False, color="orange")
plt.title("Number of sales by Product")
plt.xlabel("Product")
plt.ylabel("Total Quantity")
plt.tight_layout()
plt.show()
```

#### Insights
- Mouse has the largest amount of sales followed by Keyboard
